```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

## **Most Frequent Movie Distribution in 1990s**

```python
# Start coding here! Use as many cells as you like
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]

duration = int(movies_90s['duration'].mode()[0])

duration
```

## **Number of action movies with a duration less than 90 minutes**

There are 7 action movies with duration less than 90 minutes

### **First Way**

```python
actionMovies = movies_90s[movies_90s['genre'] == 'Action']

short_movie_count = 0

for index, row in actionMovies.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1
        
print(f"Number of action movies with a duration less than 90 minutes: {short_movie_count}")
    
```

### **Second Way**

```python
# Filter for action movies
action_movies = movies_90s[movies_90s['genre'].str.contains('Action')]

# Filter for short movies (less than 90 minutes)
short_movies = action_movies[action_movies['duration'] < 90]

# Count the number of short action movies
short_movie_count = len(short_movies)

print(short_movie_count)
```
